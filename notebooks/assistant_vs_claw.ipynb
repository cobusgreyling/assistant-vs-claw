# Assistant vs Claw

Runnable illustrations of **on-behalf-of Assistants** vs **fixed-identity Claws**.

```
from identity_models.assistant import AssistantAgent
from identity_models.claw import ClawAgent
from identity_models.fixtures import build_token_store

assistant = AssistantAgent("onboarding-agent", build_token_store())
claw = ClawAgent("product-bot", "product-bot@company.com")

for query, user_id in [("salary", "alice@co"), ("Q3 roadmap", "alice@co")]:
    a = assistant.run(user_id, query)
    c = claw.run(query, sender=user_id)
    print(f"Query: {query!r} | User: {user_id}")
    print(f"  Assistant pages: {a.pages}")
    print(f"  Claw pages:      {c.pages}")
    print(f"  Assistant audit: {a.audit_event.to_dict()}")
    print(f"  Claw audit:      {c.audit_event.to_dict()}")
    print()
```

## Common mistake

A Claw wired with personal OAuth inherits the invoker's access.

```
from identity_models.assistant import get_user_credentials
from identity_models.data import notion_search

personal = get_user_credentials("alice@co", build_token_store())
mistake = notion_search(personal, "salary")
scoped = claw.search_notion("salary")

print("Personal OAuth (mistake):", mistake)
print("Scoped Claw (correct):  ", scoped)
```
